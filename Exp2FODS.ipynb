```python
import pandas as pd
```

```python
df=pd.read_csv('/content/housing_dataset.csv')
df.head()
```

# **Handling Null Values**

```python
df.isnull().sum()
null_col=df.columns[df.isnull().any()]
print(null_col)
```

```python
df.fillna(df["Price"].median()) #filling null values in price col with its median value
```

```python
df.fillna(df["Bathrooms"].mean()) #filling the mean values in bathrooms
```

```python
df.dropna() #drop null values in the columns
```

## **Handling Outliers**

```python
#identifying outliers

q1=df['Price'].quantile(0.25)
q3=df['Price'].quantile(0.75)
IQR=q3-q1
lower_bound=q1+(1.5*IQR)
upper_bound=q3-(1.5*IQR)
print("Lower bound:",lower_bound)
print("Upper bound:",upper_bound)
```

```python
#removing the outliers from the datasets

outlier=(df[['Price']]>upper_bound) | (df[['Price']]<lower_bound)
print(outlier)
```

```python
cleaned_df=df[~outlier]
print(cleaned_df)
```

```python
#Normalizing the data on the basis of min max algo
df.columns
```

```python
# prompt: Normalize the &#39;Sqft_Living&#39; column using Min-Max scaling after removing outliers.

from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to the 'Sqft_Living' column of the cleaned DataFrame
scaler.fit(cleaned_df[['Sqft_Living']])

# Transform the 'Sqft_Living' column using the fitted scaler
cleaned_df['Sqft_Living_Normalized'] = scaler.transform(cleaned_df[['Sqft_Living']])

print(cleaned_df.head())
```
