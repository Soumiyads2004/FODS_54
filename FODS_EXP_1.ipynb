```python
sales_data=pd.DataFrame({'Months':['Jan','Feb','Mar','Jan','feb','Mar'],'Product':['A','A','A','B','B','B'],'Sales':[150,200,250,300,400,500]})
sales_data
```

```python
grouped_data = sales_data.groupby('Product')['Sales'].sum() #Group the data by &#39;Product&#39
grouped_data
```

```python
month_sales=sales_data.loc[sales_data.groupby('Product')['Sales'].idxmax()]   #Find the month with the highest sales f
month_sales
```

```python
total_sales=sales_data['Sales'].sum()
total_sales
```

```python
sales_data['sales_percentage']=(sales_data['Sales']/total_sales)*100
sales_data
```

```python
df5_1=pd.DataFrame({'OrderID':[1, 2, 3, 4, 5],'CustomerID':[101, 102, 103, 104, 101],'Product':['A','B','A','C','B']})
```

```python
df5_2=pd.DataFrame({'CustomerID':[101, 102, 103, 104],'Name':['Alice','Bob', 'Charlie', 'David'],'Location':['New York','Los Angeles','Chicago','Houston']})
```

```python
merge_df5=pd.merge(df5_1,df5_2,on='CustomerID')
merge_df5
```

```python
# prompt: Group the merged DataFrame by &#39;Location&#39; and calculate the total number of
# orders from each location.

orders_by_location = merge_df5.groupby('Location')['OrderID'].count()
orders_by_location
```
